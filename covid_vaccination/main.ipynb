#### Loading the dataset

```python
data = pd.read_csv("covid_vaccine_statewise.csv")
```

### Description of the data
1. Total Doses Administered -> Tells about the doses administered state-wise.
2. Updated On -> Dataset updated on the specific date
3. First Dose Administered  -> First Dose Administered statewise
4. Second Dose Administered -> Second Dose Administered statewise
5. Male(Doses Adminsitered) -> No. of doses administered by males.
6. Females(Doses Administered) -> No. of doses administered by females.
7. Transgender(Doses Administered) -> No. of doses administered by transgender.
8. Covaxin -> Type of Dosage taken by people.
9. Covishield -> Type of Dosage taken by people.
10. AEFI -> No. of Adverse Events Following Immunization patients

#### Dropping the columns which are having null values more than half of dataset

```python
data.drop(inplace=True,columns=["Transgender(Individuals Vaccinated)","Female(Individuals Vaccinated)","Male(Individuals Vaccinated)","60+ Years(Individuals Vaccinated)","45-60 Years(Individuals Vaccinated)","18-44 Years(Individuals Vaccinated)","60+ Years (Doses Administered)","45-60 Years (Doses Administered)","18-44 Years (Doses Administered)","Sputnik V (Doses Administered)"])
```

```python
data.isnull().sum()
```

#### Filling in the null values 

```python
data["First Dose Administered"].fillna(data["First Dose Administered"].median(),inplace=True)
data["Second Dose Administered"].fillna(data["Second Dose Administered"].median(),inplace=True)
data["Male (Doses Administered)"].fillna(data["Male (Doses Administered)"].median(),inplace=True)
data["Female (Doses Administered)"].fillna(data["Female (Doses Administered)"].median(),inplace=True)
```

```python
data.isnull().sum()
```

```python
columns_to_update =  ["First Dose Administered", "Second Dose Administered", "Male (Doses Administered)", "Female (Doses Administered)"]
for i in columns_to_update:
    data[i] = data[i].astype('int64')
```

```python
data.dtypes
```

#### Unique states of india for which the data is stored in this file.

```python
data.State.unique()
```

#### In above listed states, India is also present which is not the state. May be it contains the sum of values in all states. But it will change the values in our analysis. So we will will drop the rows containing India as State by using `drop()` function.

```python
data.drop(data[data["State"] == "India"].index, inplace=True)
data.shape
data = data.reset_index(drop=True)
data.head(5)
```

#### Used `reset_index` method to reset index of the dataframe

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Full dataset")

# Line chart  
sb.lineplot(data=data)
```

#### 1.Number of persons statewise vaccinated for first and second dose

```python
grouped_by_state = data.groupby('State')
statewiseFirst = {}
statewiseSecond = {}
for name,group in grouped_by_state:
    statewiseFirst[name] = group['First Dose Administered'].sum();
    statewiseSecond[name] = group['Second Dose Administered'].sum();
stateWiseVaccinationDosageFirst = {
    "first":statewiseFirst,
    
}
stateWiseVaccinationDosageSecond = {
"second":statewiseSecond
    
}
print(stateWiseVaccinationDosageFirst,"\n")
print(stateWiseVaccinationDosageSecond)

```

#### Graphical representation of the same

```python
sb.barplot(x=list(stateWiseVaccinationDosageFirst['first'].keys()),y=list(stateWiseVaccinationDosageFirst['first'].values()))
```

```python
sb.barplot(x=list(stateWiseVaccinationDosageSecond['second'].keys()),y=list(stateWiseVaccinationDosageSecond['second'].values()))
```

#### Pie Chart Representation of the same

```python
dosage_count = list(stateWiseVaccinationDosageSecond['second'].values())
labels = list(stateWiseVaccinationDosageSecond['second'].keys())
colors = sb.color_palette('pastel')[0:len(labels)]

#create pie chart
plt.pie(x=dosage_count, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```
