```python
import pandas as pd
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
text_file = open("./sampletext.txt",'r')
data = text_file.read()
text_file.close() #good practice
```

Tokenization 

It is the process of breaking down a stream of text into words, phrases, symbols, or any other meaningful elements called tokens

```python
tokens = nltk.tokenize.word_tokenize(data)
print(tokens)
```


Stop words

Text and document classification over social media, such as Twitter, Facebook, and so on is usually affected by the noisy nature (abbreviations, irregular forms) of the text corpuses. eg: and, the, a etc

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
# filtered_sentence = [w for w in tokens if not w in stop_words]
filtered_sentence = []

for w in tokens:
    if w not in stop_words:
        filtered_sentence.append(w)

print("Tokens Below \n")
print(tokens)
print("\nFiltered Sentence Below \n")
print(filtered_sentence)
```

POS Tagging

Part-of-speech (POS) tagging is a popular Natural Language Processing process which refers to categorizing words in a text (corpus) in correspondence with a particular part of speech, depending on the definition of the word and its context

```python
from nltk import pos_tag
pos = pos_tag(filtered_sentence)
print(pos)
```

Text Stemming

It is modifying a word to obtain its variants using different linguistic processeses like affixation (addition of affixes). For example, the stem of the word "studying" is "study", to which -ing.

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
porter=PorterStemmer()
def stemSentence(sentence):
    token_words=word_tokenize(sentence)
    token_words
    stem_sentence=[]
    for word in token_words:
        stem_sentence.append(porter.stem(word))
        stem_sentence.append(" ")
    return "".join(stem_sentence)

example = "Data science is an interdisciplinary field that uses scientific methods, processes, algorithms and systems to extract knowledge and insights from data in various forms, both structured and unstructured, similar to data mining."
print(example)
print("----------\nStemmed sentence\n----------")
x=stemSentence(example)
print(x)
```

Text lemmatization

It is the process of eliminating redundant prefix or suffix of a word and extract the base word (lemma).

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
lemmatized_text = []
print("{0:20}{1:20}".format("Word","Lemma"))
for word in filtered_sentence:
    lemmatized_text.append(wordnet_lemmatizer.lemmatize(word,pos="v"))
    print ("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word,pos="v")))
# n->nouns a->adjectives v->verbs
```

Document Frequency

```python
words_set = set()
words_set = words_set.union(set(lemmatized_text))
df = dict.fromkeys(words_set,0)
```

```python
for word in lemmatized_text:
    df[word]= df[word]+1
```

```python
df
```

Term Frequency

```python
tf = {}
corpusCount = len(lemmatized_text)
for word, count in df.items():
    tf[word] = count / corpusCount
```

```python
tf
```

Inverse Document Frequency

```python
import math
```

```python
idfDict = {}
N = len(df)
idfDict = df;
for word, val in idfDict.items():
    idfDict[word] = math.log10(N / val)
```

```python
idfDict
```

TF-IDF

```python
tfidf = {}
for word, val in tf.items():
    tfidf[word] = val * idfDict[word]
```
